```python
db_conn = sqlite3.connect("carvana_recent.db")
c = db_conn.cursor()
```

```python
c.execute(
    '''
    CREATE TABLE recon_dates(
    Vehicle_ID INTEGER,
    Recon_Week DATE,
    PRIMARY KEY(Vehicle_ID));
    '''
)
```

```python
c.execute(
    '''
    CREATE TABLE recon_work(
    Vehicle_ID INTEGER,
    Total_Recon_Cost REAL,
    PRIMARY KEY(Vehicle_ID));
    '''
)
```

```python
c.execute(
    '''
    CREATE TABLE stock_details(
    Vehicle_ID INTEGER,
    Year INTEGER,
    Make TEXT,
    Model TEXT,
    TrimLine TEXT,
    BodyStyle TEXT,
    InteriorColor TEXT,
    ExteriorColor TEXT,
    Doors INTEGER,
    OdometerValue REAL,
    AcquisitionDate DATE,
    AcquisitionSource TEXT,
    PRIMARY KEY(Vehicle_ID));
    '''
)
```

```python
recon_dates.to_sql('recon_dates', db_conn, if_exists='append', index=False)
recon_work.to_sql('recon_work', db_conn, if_exists='append', index=False)
stock_details.to_sql('stock_details', db_conn, if_exists='append', index=False)
```

```python
numbers_per_make = pd.read_sql("select A.Make, avg(A.Total_Recon_Cost) as avg_price_per_make, avg(OdometerValue) as avg_odo, count(A.Vehicle_ID) as Num_vehicles from (select * from stock_details join recon_work on stock_details.Vehicle_ID = recon_work.Vehicle_ID )A group by 1 order by avg_price_per_make desc", db_conn)
```

```python
numbers_per_make.round(2)
#observations
#Avg price spent on make with most number of vehiles reconfigured in IC is less than make with least number of vehicles reconf'd
#Chevi against land rover
```

```python
numbers_per_make.to_excel("numbers_per_make.xlsx")
```

```python
avg_odo_price_per_week = pd.read_sql('''select A.Recon_Week, avg(A.Total_Recon_Cost) as avg_price_per_make, avg(OdometerValue) as avg_odo, count(A.Vehicle_ID) as Num_vehicles from 
 (
 select * from stock_details 
 join recon_work
 on stock_details.Vehicle_ID = recon_work.Vehicle_ID
 join recon_dates
 on recon_dates.Vehicle_ID = recon_work.Vehicle_ID
 )A
 group by 1
 order by A.Recon_Week desc''', db_conn)
```

```python
avg_odo_price_per_week.to_excel("avg_odo_price_per_week.xlsx")
```

```python
avg_odo_price_per_week_per_make = pd.read_sql('''
	 select A.Recon_Week, A.Make, avg(A.Total_Recon_Cost) as avg_price_per_week, avg(OdometerValue) as avg_odo, count(A.Vehicle_ID) as Num_vehicles from (
	 select * from stock_details 
	 join recon_work
	 on stock_details.Vehicle_ID = recon_work.Vehicle_ID
	 join recon_dates
	 on recon_dates.Vehicle_ID = recon_work.Vehicle_ID
	 )A
	 group by 1,2
	 order by A.Recon_Week desc

''',db_conn)
```

```python
avg_odo_price_per_week_per_make.to_excel("avg_odo_price_per_week_per_make.xlsx")
```

```python
bodystyleimpactcost = pd.read_sql(
'''select stock_details.BodyStyle,sum(recon_work.Total_Recon_Cost) as totalcost, avg(recon_work.Total_Recon_Cost) as avgcost , count(recon_dates.Vehicle_ID) as numofvehicles from recon_work
join stock_details
on recon_work.Vehicle_ID = stock_details.Vehicle_ID
join recon_dates 
on recon_work.Vehicle_ID = recon_dates.Vehicle_ID
group by stock_details.BodyStyle''', db_conn)
```

```python
bodystyleimpactcost.to_excel("bodystyleimpactcost.xlsx")
```

```python
cost_vehicles_per_week = pd.read_sql('''select rd.Recon_Week, sum(rw.Total_Recon_Cost) as sum_of_total_recon_cost, count(rw.Vehicle_ID) as no_of_vehicles_sold
from recon_dates rd
join recon_work rw 
on rd.Vehicle_ID = rw.Vehicle_ID
group by rd.Recon_Week 
''', db_conn)
```

```python
cost_vehicles_per_week
```

```python
flag = pd.read_sql('''select A.*,
case when A.Recon_Week != '2020-07-27 00:00:00' and A.Recon_Week != '2020-08-17 00:00:00' then 'recon_cost_for_other_weeks'
	 else A.Recon_Week end as flag
from (
select rd.Recon_Week, sd.BodyStyle, rw.Total_Recon_Cost
from recon_dates rd
join recon_work rw 
on rd.Vehicle_ID = rw.Vehicle_ID
join stock_details sd
on sd.Vehicle_ID = rw.Vehicle_ID
where sd.BodyStyle = 'SUV'
order by rd.Recon_Week
)A''', db_conn)
```

```python
flag.to_excel('flag.xlsx')
```

```python
flag_sedan = pd.read_sql('''select A.*,
case when A.Recon_Week != '2020-07-27 00:00:00' then 'recon_cost_for_other_weeks_sedan'
	 else A.Recon_Week end as flag
from (
select rd.Recon_Week, sd.BodyStyle, rw.Total_Recon_Cost
from recon_dates rd
join recon_work rw 
on rd.Vehicle_ID = rw.Vehicle_ID
join stock_details sd
on sd.Vehicle_ID = rw.Vehicle_ID
where sd.BodyStyle = 'Sedan' 
order by rd.Recon_Week
)A''', db_conn)
```

```python
flag_sedan.to_excel('flag_sedan.xlsx')
```

```python
odometer_avg_cost = pd.read_sql('''select rd.Recon_Week, avg(rw.Total_Recon_Cost) as avg_total_reconf_cost, avg(sd.OdometerValue) as avg_odometer_reading
from recon_work rw
join recon_dates rd
on rd.Vehicle_ID = rw.Vehicle_ID
join stock_details sd 
on rw.Vehicle_ID = sd.Vehicle_ID
group by rd.Recon_Week
order by rd.Recon_Week asc''', db_conn)
```

```python
odometer_avg_cost.to_excel('odometer_avg_cost.xlsx')
```
